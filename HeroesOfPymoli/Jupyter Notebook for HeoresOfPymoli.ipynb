```python
###Import and Read CSV
import os
import pandas as pd

game_df = pd.read_csv('purchase_data.csv')
game_df.head(10)
```

```python
###Player Count

#The total number of players
total_players = len(game_df['SN'].unique())
print("Total players:", total_players)
```

```python
###Purchasing Analysis(Total)

#total of all item names
item_names_count = len(game_df['Item Name'].unique())

#finding average purchase price
ave_price = game_df['Price'].mean()

#total number of purchases
total_purchases = len(game_df['Price'])

#total purchases
total_revenue = game_df['Price'].sum()

# All of the data found in summary
summary_df = pd.DataFrame({"Total Unique Items": item_names_count,
                              "Average Price of Purchase": ave_price,
                              "Total Number of Purchases": total_purchases,
                              "Total Revenue": total_revenue}, index=[0])
summary_df
```

```python
########NOT DONE

###Purchasing Analysis(Gender)

gender_grouped = game_df.groupby(['Gender'])

gender_count = gender_grouped['Purchase ID'].count()

price_gender = gender_grouped['Price'].mean()

total_purchase_gender = gender_grouped['Price'].sum()

#ave_total_purchase =

table = pd.DataFrame({"Purchase Count" : gender_count, 
                     "Average Purchase Price": price_gender,
                     "Total Purchase Value": total_purchase_gender})
table
```

```python
###Age Demographics

#Bin the data and add to Dataframe
bins = [4, 9, 14, 19, 24, 29, 34, 39, 44, 49]

df_labels = ['<10', '10-14','15-19','20-24','25-29','30-34','35-39','40-44','45-49']

new_game = pd.cut(game_df['Age'], bins, labels=df_labels)

game_df['Age Range'] = pd.cut(game_df['Age'], bins, labels=df_labels)
game_df.head()
```

```python
########NOT DONE

###Purchasing Analysis(Age)

#Sort and create table
ar_group = game_df.groupby('Age Range')

ar_purchase_count = ar_group['Price'].count()

ar_mean_price = ar_group['Price'].mean()
#.map('$[:.2f]'.format)

ar_total_value = ar_group['Price'].sum()

#ar_average_per_person = 

table2 = pd.DataFrame({"Purchase Count": ar_purchase_count,
                      "Average Purchase Price": ar_mean_price,
                      "Total Purchase Value": ar_mean_price})
table2
```
